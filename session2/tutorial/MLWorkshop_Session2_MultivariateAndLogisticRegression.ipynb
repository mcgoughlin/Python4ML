### 1.2 For Loops

Loops allow us to apply the same lines of code repeatedly over each value in a data structure, allowing us to extend our code efficiently to whole collections of data. 

The 'for' loop is the most fundamental of the looping methods. Given a data structure (such as an array), the for loop will iterate through each of the structure's elements. In each iteration, the 'for' loop will create a loop variable that assumes the value of one of the structure's elements and run the loop's code. In the next iteration, the loop will reassign the loop variable's value as the next element before running the loop code again etc. This process continues until the whole data structure has been looped through.

For loops are written using the following syntax:

```
for <loop_variable> in <data_structure>:
  print("My loop variable is", <loop_variable>)
```
See an example of this below, where I print the numbers 0-5 using a series of 'print' statements, and then do the same using a for loop and a list:

````python
print(0)
print(1)
print(2)
print(3)
print(4)
print(5)

print("")

for number in [0,1,2,3,4,5]:
  print("My loop variable is",number)
````

### 1.3 Dictionaries

Introducing our next data structure: the dictionary.

Whereas in an array the most basic element is a value, the dictonary's most basic element is a key-value pair. Dictionaries are instantiated, and values are entered into the dictionary, using the following syntax:



```
our_dictionary = {} ## This line instantiates a dictionary, with the variable name 'our_dictionary'

our_dictionary['height'] = 188
our_dictionary['hair_colour'] = "brown"
our_dictionary['isMale'] = True
```

To create a new dictionary, we use the '{ }' syntax, unlike lists which use '[ ]'. To store a new key-value pair in our dictionary, we use the square brackets, enter our new key and assign the value using the '=' operator. See the code block below for and example.

In lists, arrays, and pandas dataframes, the square brackets are used for indexing - $our\_list[i]$ is used to access the $i^{th}$ element of the list $our\_list$. Dictionaries do not preserve the order of key-value pairs, so this form of indexing does not apply to dictionaries. Instead, we access and assign values to predetermined keys using the square bracket '[ ]'.

We can instantiate dictionaries with key-value pairs, and overwrite values for existing keys, using the following:


````python
second_dictionary = {'height':188, 'hair_colour':'brown','isMale':True}

print(second_dictionary)

second_dictionary['hair_colour'] = 'blonde'

print(second_dictionary)
````

See how the dictionary's stored value for the key 'hair_colour' changes between the two print statements above? This is because the value of the hair_colour key is reassigned to blonde in the above code.

Dictionaries are excellent for storing a single observation of experimental data. For example, in a simple experiment observing time, force, displacement and mass, an observation could be stored like so:

```
observation = {}
observation['time'] = 20
observation['force'] = 20000
observation['displacement'] = 5
observation['mass'] = 8472.5
```

However, we run into problems when we want to make a secondary observation. If we were to store a second observation's values in the same dictionary as our first observation, the first observation's data would be overwritten and lost. Instead, the most-used approach is to create a *list of dictionaries*. The list contains one dictionary for each observation, each observation contains variable values at each time point. 

This is done by writing each dictionary, and appending it to the list of dictionaries using the list's '.append' function:



````python
list_of_dicts = []

obs1 = {}
obs1['time'] = 1
obs1['displacement'] = 3412
list_of_dicts.append(obs1)

obs2 = {}
obs2['time'] = 2 
obs2['displacement'] = 3453
list_of_dicts.append(obs2)

obs3 = {}
obs3['time'] = 3
obs3['displacement'] = 3492
list_of_dicts.append(obs3)

print(list_of_dicts)
````

Or, using the for loop introduced earlier:

````python
list_of_dicts = []

for i in range(20):
  obs = {}
  obs['time'] = i
  obs['displacement'] = i**2 * 0.54 + 0.981*i + 14
  list_of_dicts.append(obs)

print(list_of_dicts)
````

If it isn't immediately intuitive about how useful this mode of storing data is, then fear not. A list of dictionaries is a pretty abstract concept. There are multiple reason we should store observed data as a list of dictionaries, but the most important can be shown intuitively in code.

A list of dictionaries stores several observations of data. If each dictionary's keys has the same variable name, this can be thought of as a table, where each single dictionary in a list (an observation of data) corresponds to a row, and the dictionaries keys (variable names) correspond to columns. The python package 'pandas' allows us to move from imagining this weird abstract table in our minds to actually having the table ready to manipulate at our fingertips. Simply import pandas and instantiate a 'DataFrame' like so:

````python
import pandas as pd

list_of_dicts = []

#populate our empty "list_of_dicts" with dictionaries using the for-loop below!
for i in range(20):
  obs = {}
  obs['time'] = i
  obs['displacement'] = i**2 * 0.54 + 0.981*i + 14
  list_of_dicts.append(obs)

print(list_of_dicts)

df = pd.DataFrame(list_of_dicts)

print(df.to_string(index=False))
````

Fantastic! We can see that pandas has converted our list of observations (where each observation is a dictionary in our list) to a dataset of values.

To recap our data structures: Lists, arrays, dictionaries, and pandas DataFrames are the bedrock of machine learning in Python. Lists and arrays are good for holding ordered data, dictionaries are good for holding experimental data values, and dataframes are good for providing structure to multiple observations of variables in a dataset.

Using the pandas dataset, we can apply mathematical functions to our data, access our data quickly, and plot our data. For example, the code block below plots time vs. displacement in the above table using just one line:



````python
df.plot(x='time',y='displacement')
````

### 2.1 Problem Statement

Multivariate regression is very similar to the linear regression algorithm explored in the last tutorial; we have some input and output variables, and we aim to generate a curve that best fits the relationship between the two. The *only* difference is the number of variables included in the model:

> **Multivariate regression uses two or more inputs variables to predict one or more output variables.**

In other words, multivariate regression is identical to linear regression, but can use more data to predict its outputs. As a result,  multivariate regression models can use the same simple mathematics as linear regression models - and identical optimisation mathematics, such as using the loss function and gradient descent - whilst being a much more powerful analysis tool than linear regression.

To account for this expansion in our algorithm's equation, we need to expand upon the simple linear equation shown in MLWorkshop_Week1 (shown below)...

\begin{align}
Y &= mX + C \tag{1}
\end{align}

...to include more input and output variables. This can be expressed using the following set of equations:

\begin{align}
Y_1 &= m_{1,1}X_1 + m_{1,2}X_2 + ... + m_{1,m}X_m + C_1
\end{align}
\begin{align}
Y_2 &= m_{2,1}X_1 + m_{2,2}X_2 + ... + m_{2,m}X_m + C_2
\end{align}
\begin{align}
...
\end{align}
\begin{align}
Y_n &= m_{n,1}X_1 + m_{n,2}X_2 + ... + m_{n,m}X_m + C_n \tag{2}
\end{align}

Upon first glance, the set of equations in (2) can be intimidating. However, there are really only 3 terms in each of (2)'s equations - the exact same terms used in the linear regression equation in (1):

1. $Y_i$ - The model's output. Because the model can now output multiple variables (we decide how many variables we wish our model to predict!), we must index our model's outputs with the variable $i$. $i$ can be any value between $1 <= i =< n$, where $n$ is the total number of outputs we wish to predict.

2. $m_{i,j}X_j$ - The model's input multiplied by a gradient value. Because the model can now receive multiple variables, we must index our model's inputs and gradients, too. Don't get confused between the output and input indexing! They are completely seperate and independent, and do not correspond to each other at all. We can have a multivariate model with 50 input variables (and corresponding gradients) and one output variable, and visa versa. We index our input variables with $j$ and our output variables with $i$. $j$ can be any value between $1 <= i =< m$, where $m$ is the total number of input variables we are using in our analysis.

3. $C_i$ - The model's biases. Like in the linear regression equation (1), there is only one bias value per output, so it shares the same indexing as the model's output.

The equations in (2) state that each input variable has its own gradient for each individual output variable, such that the total number of multivariate model gradients is $n$ x $m$. Further, (2) describes that each output $Y_i$ is calculated by multiplying each input variable $X_j$ by a unique gradient value $m_{i,j}$ and summed together along with bias $C$. This can be thought of as a 'weighted sum' operation, where $m_{i,j}$ is the 'weight', or relative importance, for input variable $X_j$ to output $Y_i$. This is standard terminology in data science, so from this point onwards we will refer to the $m$ term as a 'weight'.

## Bringing back the node idea again

Remember our idea of a linear regression model as a node from Workshop 1? Well, in multivariate models, this idea starts to become conceptually useful. We can now conceptualise each equation in (2) as a seperate node. This means that, for any multivariate model with $n$ output variables, our conceptual model posses $n$ nodes! Let's have a look at how this might look like in a graphic. In this model, we have 3 input variables and 2 output variables. As a result, we have two nodes in our model that each have 3 weights and one C value, known as a bias:

image.png

We know that each weight's second index, $j$, is to be multiplied by the $j^{th}$ input variable. Let's write out the relationship of each node to the input data using a graph of nodes and connections, where the strength of each connection between a node and the input data is the weight value $m_{i,j}$:

image.png

Upon first inspection, the above graph can be a little confusing, so lets unravel its meaning by tracking how a set of input data $[X_1, X_2, X_3]$ is converted into an output $Y_1$.

First, we start with some values of our input data, $[X_1, X_2, X_3]$. To generate $Y_1$, we must use the weights between the input variables and the node that outputs $Y_1$. The node for $Y_1$ is the top node, whose connections are denoted by the blue lines, and whose the $m$ values are indexed by $1$ in the first position, i.e. $m_{1,1}, m_{1,2}, m_{1,3}$. 

We multiply $X_1$ by weight $m_{1,1}$, $X_2$ by weight $m_{1,2}$, and $X_3$ by weight $m_{1,3}$, and sum these values together, before adding our bias value $C_1$. This sum total of all these values is our output, $Y_1$!

We can write this idea in a neater way using the 'sum' mathematical operator, denoted by capital sigma, $\Sigma$. The sum operator is an iterative summing procedure, like a 'for' loop, that adds together the output of each  iterations. In our context, sum is used to add each combination of $X_{j}m_{i,j}$ for all values of j. Rewriting (2) using the sum operator (3) is done like so:

\begin{align}
Y_1 &= \sum_{j=1}^{m} (m_{1,j}X_j)  + C_1
\end{align}
\begin{align}
Y_2 &= \sum_{j=1}^{m} (m_{2,j}X_j)  + C_2
\end{align}
\begin{align}
...
\end{align}
\begin{align}
Y_n &= \sum_{j=1}^{m} (m_{n,j}X_j) + C_n \tag{3}
\end{align}

Using the terminology of the for-loop to explain the sum operator, the sum operator's loop variable is located underneath the $\sum$ symbol, with it's starting index value stated (for example, $j = 1$), and the upper limit of the loop variable $m$ is stated above the sum symbol. The sum operator loops over every input variable and corresponding weight, indexed using the loop variable $j$, multiplies the input variable by its weight, and adds the result to a sum total that increase upon every iteration. The sum operator increments each value from 1 to $m$ (so $1, 2, 3..., m-1, m$), and sums the value of all $m_{i,j}X_j$.

The general form of (3), for any output $Y_i$, is computed using the following equation:

\begin{align}
Y_i &= \sum_{j=1}^{m} (m_{i,j}X_j) + C_i \tag{4}
\end{align}

Where $i$ indexes the output variables and $j$ indexes the input variables. It can be instructive at this point to compare (4) and (1) and note their similarity - again, only three terms exist: the output $Y$, the input multiplied by its weight $mX$ and the bias $C$. The only difference is the sum operator, and that we are using multiple variables - because of this, we have to match each weight to each output and input variable using convoluted annoying indexing to keep everything consistent.

In this example, we are going to use the machine-learning package called 'PyTorch' to create our multivariate regression model. To do this, we create a 'Linear' layer, and tell it how many input and output variables it can expect from our data. This can do be done in the following code block:

````python
import torch
import torch.nn as nn

X = our_dataset[['species','petal_length','petal_width']].values
Y = our_dataset[['sepal_length','sepal_width']].values

X_train,X_test = X[:int(0.8*len(X))], X[int(0.8*len(X)):]
Y_train,Y_test = Y[:int(0.8*len(Y))], Y[int(0.8*len(Y)):]
print(X.shape, Y.shape)

m = X.shape[1]
n =  Y.shape[1]

multivariate_model = nn.Linear(m,n)
````

I will write this code for you, so that you can learn how to perform the final exercise for this tutorial - which is an extension of multivariate regression - on your own. See below for the code block conducting a multivariate regression using these variables:


````python
import torch
import torch.nn as nn
import numpy as np

m = X.shape[1] # find m - the number of input variables in our analysis
n =  Y.shape[1] # find n - the number of output variables in our analysis

multivariate_model = nn.Linear(m,n) # initiate our multivariate model with a set of linear equations, for m input variables and n output variables

batch_size = 20
epochs = 50
batch_per_epoch = 50
loss_func = nn.MSELoss() # this is our MSE loss function 
opt = torch.optim.SGD(multivariate_model.parameters(),lr=0.001,momentum=0.9) # this is our 'stochastic' gradient descent optimiser

losses = []
#this is training!!!
for epoch in range(epochs): # loop the training code for as many training epochs as we state above

  for i in range(batch_per_epoch):
    opt.zero_grad() #initialise the gradient descent algorithm
    indices = np.random.randint(0,len(X_train),batch_size) # choose a random batch of input data by their indices
    x_batch, y_batch = torch.Tensor(X_train[indices]),torch.Tensor(Y_train[indices]) # create the batch of inputs and labels
    pred = multivariate_model(x_batch) # predict the batch outputs using our model
    loss = loss_func(pred,y_batch) # calculate the loss of our predictions using the labels
    loss.backward() # calculate gradient_descent parameters
    opt.step() # change the model values using the gradient descent parameters
    losses.append(float(loss)) # save the loss value to a list so we can plot the losses later

import matplotlib.pyplot as plt
plt.plot(losses) # plot our training losses
````

### 3.1 Problem Statement

Logistic Regression is very similar to multivariate regression; we have some input and output variables, and we aim to predict the relationship between the two. The *only* difference is that, in logistic regression, we are trying the predict a *discrete classification* for our output data, rather than a continuous variable. Given that we intend to ouput discrete values, our $Y_i$ value should an integer; 0, 1, or 2, and not a continous output like 1.54.  Classification is useful in medical diagnosis; for example, does our patient data indicate that they have cancer, or not? This problem requires a discrete classification, yes or no; 1 or 0.

> **Logistic regression uses one or more inputs variables to predict one or more output classifications.**

Equation (4) is shown again below to refresh our memory. (4) is the generic form of the multivariate regression model for output $i$. We use (4) in the multivariate model to predict output $i$ using all input data $X_j$.

\begin{align}
Y_i &= \sum_{j=1}^{m} (m_{i,j}X_j) + C_i \tag{4}
\end{align}

We use an near-identical equation to set up the logistic regression model, except that we add a discretising function, "sigmoid", denoted by sigma $\sigma$, that squashes the output of the multivariate model to between 0-1. See below for the generic form of the logistic regression model equation:

\begin{align}
Y_i &= \sigma\left(\sum_{j=1}^{m} (m_{i,j}X_j) + C_i\right) \tag{5}
\end{align}

Using our network concept, this is what the logistic regression model looks like in a problem with 3 input variables and 2 output classes:

image.png


Thus, logistic regression can be thought of as a multivariate regression model thats subsequently 'squashed' by the sigmoid function, into values between 0-1. The sigmoid function is defined by the graph and equation, below. The graph shows that for large input values, the sigmoid output tends to 1, and for small values sigmoid tends to 0. **You do not need to understand the details of this equation**, but it is here for your interest:

\begin{align}
\sigma(z) &= \frac{1}{1 + e^{-z}} \tag{6}
\end{align}

Our dataset is completely made up of continuous variables, and has no missing data. This is fantastic - the only preproessing steps we will have to perform will be to get rid of collinearities in our input variables, and normalisation. First step is getting rid of collinearities.


Before performing any machine learning task, we should visualise our input data. It is exceedingly difficult to visualise 30 input variables in one graph. Instead of plotting one graph, we are going to visualise our data with a 30x30 correlation matrix that will reveal the correlations between all input variables in our dataset, otherwise known as collinearities. **Run the code block below to plot this correlation matrix**

````python
corr_matrix = bc_dataset.copy().drop(['id','diagnosis'],axis=1).corr()
plt.imshow(corr_matrix, vmin=0, vmax = 1)
print("Column names for our input data:",corr_matrix.columns.values,"\n")
````

In the above plot, yellow corresponds to perfect correlation and navy blue corresponds to no correlation at all. We expect the top-left to bottom-right diagonal to be perfectly correlated - this diagonal represents the self-correlation of each variable which should always be perfect; 1. We see this, which indicates our correlation matrix is valid.

Our dataset reveals to us that there is a fair bit of collinearity between different variables, with high correlations found outside of the self-correlation diagonal. This is not ideal - it indicates that the same variables are effectively being measured multiple times, and it could invoke error in our model which aims to find predictive relationships using both variables simultaneously. If we look at the printed column names, we see that multiple version of "radius", "area", "perimeter" etc. are measured - this confirms that the same variables causing collinearities are actually the same variables being measured multiple times. Let's get rid of this collinearity by ridding ourselves of this unnecessary data - this occurs most egregiously at column positions: 2-3, 12-13, 22-23, so we will drop these columns from our dataset.

````python
import matplotlib.pyplot as plt

noncoll_bc = bc_dataset.copy().drop(['id','diagnosis'],axis=1)
cols = noncoll_bc.columns

noncoll_bc = noncoll_bc.drop([*cols[2:4].values,*cols[12:14].values,*cols[12:24].values],axis=1)
plt.imshow(noncoll_bc.corr(), vmin=0, vmax = 1)
print("Column names for our input data:",noncoll_bc.columns.values,"\n")
````

This correlation grid is much more acceptable than the previous! This is visible by the lack of strong collinearity between input variables compared to the previous matrix. 

Now, we must split our data into train and test sets using the 80%:20% splitting ratio. However, our dataset contains an unbalanced number of malignant and benign cases; as stated above, our dataset contains "212 malignant cases and 357 benign." Unbalanced datasets can cause unwanted biases in our classification model, so we don't want to simply sample 80% of our total data to train our algorithm. There are many ways to deal with an unbalanced dataset - we will use the least sophisticated method.

We will select the class with the least representation in a our dataset - in this case, malignant cases, and sample 80% of these cases. In our case, 80% of 212 is 170 cases. We will then sample an equal number - 170 - of benign cases from our dataset - the combination of these two samples will create a training dataset of 340 cases. Our test dataset will contain the rest, 20%, of the malignant cases, and the rest of the benign cases.

**Run the code below to perform this split**.

````python
our_data = noncoll_bc
our_data['diagnosis'] = bc_dataset['diagnosis']

train_frac = 0.8
n_samples = int(maligs*train_frac) # n_samples from each class is 80% of the total number of malignant casess

pos_train = our_data[our_data['diagnosis']=='M'].sample(n_samples) ## where the diagnosis is "M" (malignant) in our data, sample n_samples. In other words, sample 170 malignant cases.
neg_train = our_data[our_data['diagnosis']!='M'].sample(n_samples) ## where the diagnosis is "B" (benign) in our data, sample n_samples. In other words, sample 170 benign cases.
train_df = pd.concat([pos_train,neg_train]) ## combine the above samples into one dataframe
#shuffle rows of train_df
train_df = train_df.sample(frac=1.0)
train_df = train_df.sample(frac=0.8).reset_index(drop=True) ## shuffle the order of the samples so they lose their row-wise structure

test_df = our_data.merge(train_df,how='outer',indicator=True) ## these two lines sample the rest of our_data that isn't included in the "train_df" dataframe
test_df = test_df[test_df['_merge']!='both'].drop('_merge',axis=1).sample(frac=1)

X_train = train_df.copy().drop('diagnosis',axis=1).values ## seperate the input variables from the output classification labels
Y_train = train_df['diagnosis'].values ## seperate the output classification labels
Y_train = np.where(Y_train=='M',1,0) ## Convert 'M'/'B' label to the numeric 1/0 values

X_test = test_df.drop('diagnosis',axis=1).values ## same as above for test set
Y_test = test_df['diagnosis'].values
Y_test = np.where(Y_test=='M',1,0)
````

Finally, we normalise each column of the training data seperately using the code block below. **We cover normalisation and preprocessing in the next tutorial. For now, it is enough to say that our normalised variables make the learning process much faster and more stable**.

We should *not* normalise test and training data together. Testing data is presumed to be unseen in our validation, which means that we should not factor any part of its data into our analysis until after training is completed. 

***Run the code block below to generate our normalised data***:

````python
X_norm_values = {}
X_train_normed = X_train.copy()

for index in range(len(X_train[0])):
  column = our_data.columns[index]
  data = X_train.copy()[:,index]
  X_norm_values[column] = [data.mean(),data.std()]
  X_train_normed[:,index] = (data - data.mean())/data.std()
````
